```python
import matplotlib.pyplot as plt

resV = []
resY = []
resT = []

for i in range(250):
  resTempV = 10
  resTempY = 0
  resTempT = 0
  if i == 0:
    resV.append(10)
    resY.append(0)
    resT.append(0)
  if i>0:
    for x in range(i,0, -1):
      resTempV = resTempV + (0.01*(-9.81))
      resTempY = resTempY + (0.01*resV[x-1])
      resTempT = resTempT + 0.01
    resV.append(resTempV)
    resY.append(resTempY)
    resT.append(resTempT)

plt.figure()
plt.plot(resT, resY)
plt.xlabel("Função T")
plt.ylabel("Função Y")
plt.title("Valores de Y em função de T")
plt.show()
print()

plt.figure()
plt.plot(resT, resV)
plt.xlabel("Função T")
plt.ylabel("Função V")
plt.title("Valores de V em função de T")
plt.show()
```
